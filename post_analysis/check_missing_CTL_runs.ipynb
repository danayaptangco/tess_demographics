```python
import pandas as pd
from pathlib import Path

ctl_path = Path("take_two_stellar_params_CTL.csv")
tls_path = Path("take_two_tls_results_combined.csv")

def get_tics(path):
    df = pd.read_csv(path)
    preferred = ["TIC", "tic", "TIC_ID", "tic_id", "TICID", "ticid", "id"]
    col = next((c for c in preferred if c in df.columns), None) or \
          next((c for c in df.columns if "tic" in c.lower()), None)
    if col is None:
        raise ValueError(f"No TIC-like column found in {path.name}. Columns: {list(df.columns)}")
    tics = set(
        df[col]
        .dropna()
        .astype(str)
        .str.strip()
        .str.replace(r"\.0+$", "", regex=True)
        .str.replace("TIC ", "")
    )
    return col, tics

ctl_col, ctl_tics = get_tics(ctl_path)
tls_col, tls_tics = get_tics(tls_path)

common = ctl_tics & tls_tics
ctl_only = ctl_tics - tls_tics
tls_only = tls_tics - ctl_tics

print(f"{ctl_path.name}: {len(ctl_tics)} unique TICs (column: {ctl_col})")
print(f"{tls_path.name}: {len(tls_tics)} unique TICs (column: {tls_col})")
print(f"Common TICs: {len(common)}")
print(f"Only in {ctl_path.name}: {len(ctl_only)}")
print(f"Only in {tls_path.name}: {len(tls_only)}")
print("\nExamples only in CTL:", sorted(ctl_only)[:20])
print("Examples only in TLS:", sorted(tls_only)[:20])


#they probably just didn't get run and I just need to run them again
```
