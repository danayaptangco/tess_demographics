```python
#10GYR SOLAR

#Notes. Radii of planets, in R_oplus. Column 1 is incident flux on the planet, relative to the solar constant. Column 2 is the total planet mass in M_oplus. Otherwise, column headers indicate the fraction of a planet's mass in the H/He envelope.

from scipy.interpolate import interp1d
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



columns = ['Flux (F_oplus)', 'Mass (M_oplus)', '0.01%', '0.02%', '0.05%', '0.1%', '0.2%', '0.5%', '1%', '2%', '5%', '10%', '20%']
low_mass_planet_df = pd.read_csv('./Lopez_Fortney_data/low_mass_planet_R_10GYR_solar_metallicity.txt', delim_whitespace=True, skiprows=5, names=columns)

# Plot the data
fig, ax = plt.subplots(figsize=(10, 6))

low_mass_planet_df_1000 = low_mass_planet_df[low_mass_planet_df['Flux (F_oplus)'] ==1000]
low_mass_planet_df_10 = low_mass_planet_df[low_mass_planet_df['Flux (F_oplus)'] ==10]
low_mass_planet_df_pt1 = low_mass_planet_df[low_mass_planet_df['Flux (F_oplus)'] ==.1]

for column in columns[2:]:
    ax.scatter(low_mass_planet_df_1000['Mass (M_oplus)'], low_mass_planet_df_1000[column], label=f'{column} (1000F)')
    ax.scatter(low_mass_planet_df_10['Mass (M_oplus)'], low_mass_planet_df_10[column], label=f'{column} (10F)')
    ax.scatter(low_mass_planet_df_pt1['Mass (M_oplus)'], low_mass_planet_df_pt1[column], label=f'{column} (0.1F)')

ax.set_xlabel('Mass (M_earth)')
ax.set_ylabel('Radius (R_earth)')
ax.set_title('Radius vs Mass for Different Hydrogen Fractions')
ax.legend(title='Hydrogen Fraction', bbox_to_anchor=(1.05, 1), loc='upper left')
ax.grid(True)
#ax.set_ylim(1.5,4)

plt.show()

```

```python
# Plot only points with radius between 2 and 4 R_earth
# and mass between 5.5 and 13 M_earth
fig, ax = plt.subplots(figsize=(10, 6))

for flux_label, df_flux in [("0.1F", low_mass_planet_df_pt1),
                            ("10F", low_mass_planet_df_10),
                            ("1000F", low_mass_planet_df_1000)]:
    mass = df_flux['Mass (M_oplus)']
    for col in columns[2:]:
        radius = df_flux[col]
        mask = (radius >= 2.0) & (radius <= 4.0) & (mass >= 5.5) & (mass <= 13.0)
        ax.scatter(
            mass[mask],
            radius[mask],
            s=20,
            label=f"{col} ({flux_label})"
        )

ax.set_xlabel('Mass (M_earth)')
ax.set_ylabel('Radius (R_earth)')
ax.set_title('Radius vs Mass (2 <= Radius <= 4, 5.5 <= Mass <= 13)')
ax.grid(True)
ax.set_ylim(2, 4)
ax.set_xlim(5.5, 13)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
plt.show()
```

```python
mass_col = 'Mass (M_oplus)'
flux_col = 'Flux (F_oplus)'

mass_mask = (low_mass_planet_df[mass_col] >= 5.5) & (low_mass_planet_df[mass_col] <= 13.0)
subset = low_mass_planet_df.loc[mass_mask, [flux_col, mass_col] + columns[2:]]

matches = []
for _, row in subset.iterrows():
    for env_col in columns[2:]:
        r = row[env_col]
        if 1.8 <= r <= 4.0:
            matches.append({
                flux_col: row[flux_col],
                mass_col: row[mass_col],
                'Envelope Fraction': env_col,
                'Radius (R_oplus)': r
            })

matches_df = pd.DataFrame(matches).sort_values([flux_col, mass_col, 'Radius (R_oplus)'])
print(matches_df.to_string(index=False))
env_unique = sorted(matches_df['Envelope Fraction'].unique(), key=lambda x: float(x.rstrip('%')))
print(env_unique)
```
