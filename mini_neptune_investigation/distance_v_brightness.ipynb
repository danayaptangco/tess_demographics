```python
import pandas as pd

import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('TOI_Mar2025.csv', skiprows=71)

rad_max = 4
rad_min = 1.5

if 'tfopwg_disp' in df.columns:
    df = df[df['tfopwg_disp'] != 'FP']

df = df[(df['pl_rade'] < rad_max) & (df['pl_rade'] > rad_min)]


# Plot st_dist by t_eff
plt.figure(figsize=(10, 6))
plt.scatter(df['st_teff'], df['st_dist'], alpha=0.7)
plt.title('Stellar Distance (st_dist) by Effective Temperature (t_eff)')
plt.xlabel('Effective Temperature (K)')
plt.ylabel('Stellar Distance (pc)')
plt.xlim(2500,8000)
plt.ylim(0,800)
plt.grid(True)
plt.show()
```
